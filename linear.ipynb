```python
#month抽出
def mon(month):
    for i in range(len(month)):
        if month[i].find('03/') > -1:
            month[i]=3
        elif month[i].find('04/') > -1:
            month[i]=4
        elif month[i].find('05/') > -1:
            month[i]=5
        elif month[i].find('06/') > -1:
            month[i]=6
        elif month[i].find('07/') > -1:
            month[i]=7
        elif month[i].find('08/') > -1:
            month[i]=8
        elif month[i].find('09/') > -1:
            month[i]=9
        elif month[i].find('10/') > -1:
            month[i]=10
        elif month[i].find('11/') > -1:
            month[i]=11
        else:
            month[i]=12
    return month
train_month=mon(np.array(train["gameday"]))
test_month=mon(np.array(test["gameday"]))
train=pd.concat([train,pd.DataFrame(train_month)],axis=1)
test=pd.concat([test,pd.DataFrame(test_month)],axis=1)
train=train.rename(columns={0:"month"})
test=test.rename(columns={0:"month"})
```

```python
#homepointsをj1,j2の区別をありに
def home_pints(df_all):
    homepoints=[0]*len(df_all)
    for i in range(len(df_all)):
        if df_all["year"][i]==2012:
            if df_all["stage"][i]=='Ｊ１':
                homepoints[i]=df_all["home points"][i]-25
            else:
                homepoints[i]=df_all["home points"][i]-71
        elif df_all["year"][i]==2013:
            if df_all["stage"][i]=='Ｊ１':
                homepoints[i]=df_all["home points"][i]-30
            else:
                homepoints[i]=df_all["home points"][i]-78
        else:
            if df_all["stage"][i]=='Ｊ１':
                homepoints[i]=df_all["home points"][i]-21
            else:
                homepoints[i]=df_all["home points"][i]-80
    return homepoints
train_homepoints=home_pints(train)
test_homepoints=home_pints(test)
train=pd.concat([train,pd.DataFrame(train_homepoints)],axis=1)
test=pd.concat([test,pd.DataFrame(test_homepoints)],axis=1)
train=train.rename(columns={0:"home/points"})
test=test.rename(columns={0:"home/points"})
```

```python
#awaypointsをj1,j2の区別ありに
def away_pints(df_all):
    awaypoints=[0]*len(df_all)
    for i in range(len(df_all)):
        if df_all["year"][i]==2012:
            if df_all["stage"][i]=='Ｊ１':
                awaypoints[i]=df_all["away points"][i]-25
            else:
                awaypoints[i]=df_all["away points"][i]-71
        elif df_all["year"][i]==2013:
            if df_all["stage"][i]=='Ｊ１':
                awaypoints[i]=df_all["away points"][i]-30
            else:
                awaypoints[i]=df_all["away points"][i]-78
        else:
            if df_all["stage"][i]=='Ｊ１':
                awaypoints[i]=df_all["away points"][i]-21
            else:
                awaypoints[i]=df_all["away points"][i]-80
    return awaypoints

train_awaypoints=away_pints(train)
test_awaypoints=away_pints(test)
train=pd.concat([train,pd.DataFrame(train_awaypoints)],axis=1)
test=pd.concat([test,pd.DataFrame(test_awaypoints)],axis=1)
train=train.rename(columns={0:"away/points"})
test=test.rename(columns={0:"away/points"})
```

```python
#天気を変換
def rain(df_all):
    weather=[0]*len(df_all)
    for i in range(len(df_all)):
        if df_all["main weather"][i]=="雨":
            weather[i]=1
        else:
            weather[i]=0
    return weather

train_weather=rain(train)
test_weather=rain(test)
train=pd.concat([train,pd.DataFrame(train_weather)],axis=1)
test=pd.concat([test,pd.DataFrame(test_weather)],axis=1)
train=train.rename(columns={0:"rain"})
test=test.rename(columns={0:"rain"})
```

```python
city = ['北海道','宮城県','福島県', '埼玉県', '千葉県', '東京都', 
        '神奈川県', '新潟県', '静岡県', '愛知県','京都府', '大阪府', 
        '兵庫県','岡山県','広島県','福岡県', '熊本県']

def urban(prefecture):
    prefecturecount=[0]*len(prefecture)
    for i in range(len(prefecture)):
        for name in city:
            if prefecture[i]==name:
                prefecturecount[i]=1
                break
    return prefecturecount

train_prefecture=urban(np.array(train["prefecture"]))
test_prefecture=urban(np.array(test["prefecture"]))
train=pd.concat([train,pd.DataFrame(train_prefecture)],axis=1)
test=pd.concat([test,pd.DataFrame(test_prefecture)],axis=1)
train=train.rename(columns={0:"urban"})
test=test.rename(columns={0:"urban"})
```
